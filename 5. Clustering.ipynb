```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
features = df.drop('Species', axis = 1)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(features)
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(range(1, 11), wcss, marker ='o', color = 'red')
plt.title('Optimal number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 100, random_state = 0)
y_means = kmeans.fit_predict(data_scaled)
```

```python
plt.scatter(data_scaled[y_means == 0, 0], data_scaled[y_means == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(data_scaled[y_means == 1, 0], data_scaled[y_means == 1, 1], s = 100, c = 'red', label = 'Iris-versicolour')
plt.scatter(data_scaled[y_means == 2, 0], data_scaled[y_means == 2, 1], s = 100, c = 'blue', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'orange', label = 'centroid')
plt.legend()
plt.show()
```
