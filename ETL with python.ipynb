```python
Q1 = df1['Fare'].quantile(0.25)
Q3 = df1['Fare'].quantile(0.75)
IQR = Q3 - Q1
width = 5
lower = Q1 - (width * IQR)
upper = Q3 + (width * IQR)

# Remove outliers
df1 = df1[(df1['Fare'] >= lower) & (df1['Fare'] <= upper)]

# Now the 'Fare' column contains data without outliers
print(df1.head())
```
